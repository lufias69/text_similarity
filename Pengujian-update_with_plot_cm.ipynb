```python
import pandas as pd
from sklearn.metrics import classification_report
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
import datetime
from sklearn.metrics import confusion_matrix
from Nawa import plot_confusion_matrix as pl

waktu = str(datetime.datetime.now()).replace(":","-")

list_data = 'dice_tf-idf jaccard_tf-idf cosine_tf-idf dice_tf-idf-df jaccard_tf-idf-df cosine_tf-idf-df dice_df jaccard_df cosine_df'
list_data = list_data.split()
```

```python
path = "hasil/pelabelan/"
nama_file = "hasil_transform_2019-04-20 21-25-55.543165.xlsx" #masukkan juga formatnya misal: nama_file.xlsx
f = pd.read_excel(path+nama_file, sheet_name = 'Sheet1')
f2 = pd.read_excel("data/jawaban_clean_jawaban buat label_1.xlsx", sheet_name = 'Sheet2')#hasil asli
f2 = f2[:100]

y_true = f2['Column4'].tolist()
target_names = ['sangat mirip', 'mirip', 'tidak mirip','sangat tidak mirip']
metode = list()
accuracy = list()
precision = list()
recall = list()
f1_score = list()

for x,i in enumerate(list_data):
    if x == 0:
        print("")
        print("")
    else:
        print("")
        
    data = f[i].tolist()
    y_pred = data
    metode.append(i)
    accuracy.append(accuracy_score(y_true, y_pred, normalize=True))
    hasil = list(precision_recall_fscore_support(y_true, y_pred, average='micro')) #weighted
    precision = round(hasil[0], 2)
    recall = round(hasil[1],2)
    f1_score = round(hasil[2],2)
    print(i.upper())
    #print(confusion_matrix(y_true, y_pred, labels=target_names))
    pl.plot_confusion_matrix(y_true, y_pred, classes=target_names,
                      title='Confusion matrix, without normalization')
    pl.plt.show()
    
    print("-"*60)
    print(classification_report(y_true, y_pred, target_names=target_names))
    print("="*60)
    print("")

print ("Note:")
print("""The reported averages include micro average (averaging the total true positives, 
false negatives and false positives), macro average (averaging the unweighted mean per label), 
weighted average (averaging the support-weighted mean per label) 
and sample average (only for multilabel classification)
link -> https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html""")
    
dic2 = {
    "metode":metode ,
    "accuracy":accuracy,
    "precision":precision,
    "recall":recall,
    "f1_score":f1_score,
}
df2 = pd.DataFrame.from_dict(dic2)
#hasil tersimpan di -> hasil\pengujian -> lihat file paling terakhir
df2.to_excel('hasil\pengujian\hasil_pengujian_update1'+waktu+'.xlsx')
df2
```
